```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# 폰트 설정

from matplotlib import font_manager, rc
font_path = "C:/Windows/Fonts/gulim.ttc"
font = font_manager.FontProperties(fname=font_path).get_name()
rc("font", family=font)
```

```python
df_wine = pd.read_csv("../csv/eda/wines.csv")
```

```python
df_review = pd.read_csv("../csv/eda/reviews.csv")
```

```python
sns.set_context("notebook")
plt.figure(figsize=(12, 6))
t = df_wine.country.value_counts().to_frame()[:20]
g = sns.barplot(x=t["count"], y=t.index, data=t)
g.set_title("생산국별 와인 분포", size=20)
```

```python
plt.figure(figsize=(12, 6))
t = df_wine[["country", "count"]].groupby("country").sum().sort_values("count", ascending=False)[:20]
g = sns.barplot(x=t["count"], y=t.index, data=t)
g.set_title("생산국별 평가 수", size=20)
```

```python
plt.figure(figsize=(12, 6))
t = df_wine.groupby("country")["price"].mean().sort_values(ascending=False).to_frame()[:20]
g = sns.pointplot(x=t["price"], y=t.index, data=t)
g.set_title("생산국별 평균 가격", size=20)
```

```python
plt.figure(figsize=(12, 6))
g = sns.stripplot(x="price", y="country", data=df_wine, jitter=True, dodge=True, marker="o", alpha=0.7, color="darkred", size=1)
g.set_title("생산국별 가격 분포", size=20)
```

```python
plt.figure(figsize=(12, 6))
t = df_wine[["country", "price"]].groupby("country").max().sort_values("price", ascending=False)[:20]
g = sns.barplot(y=t.index, x=t["price"], data=t)
g.set_title("생산국별 최대 와인 가격", size=20)
```

```python
plt.figure(figsize=(12, 6))
g = sns.boxplot(x="user_rating", y="country", data=df_wine)
g.set_title("생산국별 평점 분포", size=20)
```

```python
plt.figure(figsize=(12, 6))
t = df_wine[["acidity", "intensity", "sweetness", "price", "count", "user_rating"]].corr()
g = sns.heatmap(t, annot=True)
```

```python
plt.figure(figsize=(12, 6))
g = sns.regplot(x="user_rating", y="price", data=df_wine, line_kws={"color":"black"}, x_jitter=True, fit_reg=True, color="red")
```
